```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["school_name"].unique())

total_students = len(school_data_complete["Student ID"].unique())

budget = sum(school_data_complete["budget"].unique())

average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()

m = school_data_complete["math_score"]
r = school_data_complete["reading_score"]
math_p = (m[m > 69].count() / len(m)) * 100
reading_p = (r[r > 69].count() / len(r)) * 100

mp = school_data_complete["math_score"] >= 70
rp = school_data_complete["reading_score"] >= 70
overall_passing = (len(school_data_complete[mp & rp]) / total_students) * 100
```

```python
direct_df = pd.DataFrame(
    {"Total Schools":[total_schools],
    "Total Students":[total_students], 
    "Total Budget":[budget],
    "Average Math Score":[average_math],
    "Average Reading Score":[average_reading],
    "% Passing Math":[math_p], 
    "% Passing Reading":[reading_p], 
    "% Overall Passing":[overall_passing]}
)

#district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)

direct_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School type
school_type = school_data_complete.groupby("school_name")["type"].unique()
#print(school_type)

#Total students
total_students = school_data_complete.groupby("school_name")["student_name"].count()
#print(total_students)

#Total school budget
total_budget = school_data.groupby("school_name")["budget"].sum()
#print(total_budget)

#Per student budget
per_student_budget = total_budget/total_students
#print(per_student_budget)

#Average math score
math_score = school_data_complete.groupby("school_name")["math_score"].mean()
#print(math_score)

#Average reading score
reading_score = school_data_complete.groupby("school_name")["reading_score"].mean()
#print(reading_score)

#% passing math (the percentage of students who passed math)
math_passing = ((school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").math_score.count())/total_students)*100
#print(math_passing)

#% passing reading (the percentage of students who passed reading)
reading_passing = ((school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").reading_score.count())/total_students)*100
#print(reading_passing)

#% overall passing (the percentage of students who passed math AND reading)
overall_passing = ((school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name").math_score.count())/total_students)*100
#print(overall_passing)

district_summary_df = pd.DataFrame(
    {
    "School Type": school_type,
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Budget Per Student": per_student_budget,
    "Average Math Score": math_score,
    "Average Reading Score": reading_score,
    "% Passing Math": math_passing,
    "% Passing Reading": reading_passing,
    "% Passing Overall":overall_passing}
)
#district_summary_df["Budget Per Student"] = district_summary_df["Budget Per Student"].astype(float).map("${:,.2f}".format)
#district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)

district_summary_df
```
